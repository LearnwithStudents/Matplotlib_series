### Basic plots


```python
import matplotlib.pyplot as plt
```

```python
# x values,y values
# Raw data /Dummy data
ages=[x for x in range(15,25)]

python_dev_salaries=[15000,20000,25000,30000,35000,40000,45000,50000,60000,70000]

JS_dev_salaries=[5000,10000,15000,20000,25000,30000,37000,43000,49000,61000]

Java_dev_salaries=[7000,11000,17000,24000,29000,33000,47000,49000,52000,66000]
```

```python
plt.style.use('seaborn')
plt.plot(ages,python_dev_salaries,label='Python',color='blue',linestyle='dotted',linewidth=3)
plt.plot(ages,JS_dev_salaries,label='JavaScript',color='red',linestyle='dashed',linewidth=3)
plt.plot(ages,Java_dev_salaries,label='Java',color='green')


plt.title('Salaries of the dev based on their Ages')
plt.xlabel('Age of the Developers')
plt.ylabel('Salaries in INR')
plt.legend()
```
