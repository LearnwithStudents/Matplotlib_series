# Bar Chart's in Matplotlib

```python
import matplotlib.pyplot as plt




age_of_developers=[x for x in range(25,30)]
salaries_in_2021=[23,27,35,43,51]#in thousands
salaries_in_2022=[27,34,39,44,57]#in thousands
new_ages=[x+0.4 for x in range(25,30)]
plt.figure(figsize=(10,6))
plt.title('Salaries based on their ages')
plt.xlabel('Age of the developer')
plt.ylabel('salaries in INR(Thousands)')
plt.bar(age_of_developers,salaries_in_2021,width=0.4,color='maroon',label='2021',edgecolor='black')
plt.bar(new_ages,salaries_in_2022,width=0.4,color='grey',label='2022',edgecolor='black')
plt.legend()
```
