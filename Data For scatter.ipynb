# Scatter Plots

Example: Google released a product and they sent two survey teams(survey_1 team and survey_2 team) to get 20 users experience of their product.

```python
import matplotlib.pyplot as plt

#Both survey teams will be on same people(product_users)
#survey values are based on the users satisfaction
#All values are inbetween 1-10 numbers
survey_1=[2,4,3,6,3,8,9,1,3,7,1,2,4,9,8,9,1,4,7,3]
survey_2=[3,4,8,1,3,9,6,4,7,7,9,4,7,9,1,8,1,8,4,6]


#These values represent that the users will continue to use Google product/not
values=[7,3,6,2,9,7,4,5,7,2,5,8,9,1,4,6,9,1,9,5]


#multplied values with 10 for easy visualization
dup=[]
for i in range(len(values)):
    tmp=values[i]*50
    dup.append(tmp)
```

```python
fig=plt.gcf()
fig.set_size_inches(19.5,10.5)

plt.title('Google reviews',fontdict={'size':30},pad=30)
plt.xlabel('survey team 1',fontdict={'size':15,'color':'#666'})
plt.ylabel('survey team 2',fontdict={'size':15,'color':'#666'})
plt.scatter(survey_1,survey_2,s=dup,c=dup,cmap='Blues',edgecolor='black',alpha=0.75)
# plt.colorbar()
cbar=plt.colorbar()
cbar.set_label('will the users continue to use product')
```
